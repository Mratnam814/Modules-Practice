```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
#barplot
res = sns.barplot(df['sepal_length'],df['sepal_width'])
plt.show()
```

```python
res=sns.barplot(x="sepal_length",y="sepal_width",data=df)
```

```python
res=sns.barplot(x="sepal_length",y="sepal_width",data=df,color="yellow")
plt.show()
```

```python
res = sns.barplot(df['sepal_length'],df['petal_length'],palette='Set2')
```

```python
#countplot
sns.countplot(x="sepal_length",data=df,palette="Set1")
```

```python
sns.countplot(y="sepal_width",data=df,palette="rocket")
```

```python
sns.countplot(x='sepal_length',hue="sepal_width",data=df,palette="Set1")
```

```python
#distplot
sns.distplot(df.sepal_length,bins=10,color='g')
plt.show()
```

```python
#Heatmap
sns.heatmap(df.corr(),cbar=True,linewidths=0.5)
```

```python
df['species'].value_counts()
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",data=df)
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",data=df,hue="species",size="species")
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="species",palette="Set1")
```

```python
#lmplot
sns.lmplot(x="petal_length",y="petal_width",data=df)
```

```python
sns.lmplot(x="petal_length",y="petal_width",hue="species",data=df,markers=["o","*","^"])
```
