```python
#line plot
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("First line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color="orange",linestyle=":",linewidth=3)
plt.title("First line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color="g",linestyle=":",linewidth=2)
plt.plot(x,y2,color="r",linestyle="-",linewidth=3)

plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color="yellow")

plt.subplot(1,2,2)
plt.plot(x,y2,color="orange")

plt.show()
```

```python
#BarPlot
student={"bob":87,"Matt":56,"Sam":27}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.title("Marks of students")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#Horizonatal bar plot
student={"bob":87,"Matt":56,"Sam":27}
names=list(student.keys())
values=list(student.values())
plt.barh(names,values,color="green")
plt.title("Marks of students")
plt.xlabel("Marks")
plt.ylabel("Names")
plt.grid(True)
plt.show()
```

```python
#Scatterplot
```

```python
x=list(range(10,100,10))
y=[4,5,6,67,7,78,8,7,9]
plt.scatter(x,y)
plt.show()
```

```python
x=list(range(10,100,10))
y=[4,5,6,67,7,78,8,7,9]
plt.scatter(x,y,marker="*",c="orange",s=500)
plt.show()
```

```python
x=list(range(10,100,10))
a=[4,5,6,67,7,78,8,7,9]
b=[2,3,4,5,23,5,6,8,4]
plt.scatter(x,a,marker="*",c="orange",s=500)
plt.scatter(x,b,marker=".",c="yellow",s=100)
plt.show()
```

```python
x=list(range(10,100,10))
a=[4,5,6,67,7,78,8,7,9]
b=[2,3,4,5,23,5,6,8,4]
plt.subplot(1,2,1)

plt.scatter(x,a,marker="*",c="orange",s=500)
plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="yellow",s=500)
plt.show()
```

```python
#Histogram
data=[1,3,3,3,3,5,6,5,3,8,9,6,5]
plt.hist(data,color="g",bins=12)
plt.show()
```

```python
#BoxPLot(5 number summary)
l=[1,2,3,4,5,6,7,8,9]
l1=[1,2,3,4,5,4,3,2,1]
l2=[6,7,8,9,8,7,6,5,4]
data=list([l,l1,l2])
plt.boxplot(data)
plt.show()
```

```python
plt.violinplot(data,showmedians=True)
plt.show()
```

```python
#piechart
fruit=["apple","orange","mango","guava"]
quantity=[53,43,12,97]
plt.pie(quantity,labels=fruit,autopct="%0.2f%%")
plt.show()
```

```python
#Dough-NutChart
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=["w"],radius=1)
plt.show()
```
